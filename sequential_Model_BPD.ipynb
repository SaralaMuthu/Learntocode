```python
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
import numpy as np
import pandas as pd
# load the dataset
dataset = pd.read_csv("/content/bpdfull_kerastest_noid.csv",delimiter=',',engine='python')
# split into input (X) and output (y) variables
X = dataset.iloc[:,:dataset.shape[1]-1]
y = dataset.iloc[:,-1]
print(X)
print(y)
```

```python
# define the keras model
model = Sequential()
model.add(Dense(12, input_shape=(54675,), activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset
model.fit(X, y, epochs=100, batch_size=10)
_,accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
predictions = (model.predict(X) > 0.8).astype(int)
# summarize the first 5 cases
for i in range(5):
  print(predictions[i])
 #print('(expected %d)' , predictions[i])
```
