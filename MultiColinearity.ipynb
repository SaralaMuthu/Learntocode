MultiColinearity

```python
import pandas as pd
```

```python
import statsmodels.api as sm
df_adv = pd.read_csv('Advertising.csv', index_col=0)
X = df_adv[['TV', 'radio','newspaper']]
y = df_adv['sales']
df_adv.head()
```

```python
X = sm.add_constant(X)
```

```python
X
```

```python
## fit a OLS model with intercept on TV and Radio

model= sm.OLS(y, X).fit()
```

```python
model.summary()
```
